## *Conditional Statements*

*Conditional Statement* merupakan statement yang mengatur aliran program dengan alur keputusan. Keputusan yang diambil tergantung pada kondisi yang terjadi. Kata kunci `if` yang berarti "jika" digunakan pada *Conditional Statement* tersebut.

<br>
Sebagai contoh, Siswa akan lulus jika nilainya sama atau lebih dari 75. <br><br>
Dengan kata lain : <br>
<b>jika</b> [nilai siswa sama atau lebih dari 75] <b>maka</b> [siswa akan lulus] <br> <br>
Pada contoh diatas : <br>
*   [nilai siswa sama atau lebih dari 75] --> adalah sebuah kondisi <br>
*   [siswa akan lulus] --> adalah sebuah keputusan/aksi yang dilakukan jika kondisi terpenuhi



### `if`

Mari kita buat contoh program python untuk kasus di atas.

```python
# Contoh If

nilai = 85

if nilai >= 75 :
  print("Siswa Lulus !")
```

Pada kasus di atas, siswa mendapat nilai 85. Selanjutnya sebuah keputusan akan ditentukan dengan ketentuan "jika siswa mendapat nilai sama atau lebih dari 75, maka siswa dinyatakan lulus". Karena siswa mendapat nilai lebih dari 75, maka siswa dinyatakan lulus.

Pada contoh di atas, [nilai >= 75] merupakan sebuah kondisi dan [print("Siswa Lulus !")] merupakan sebuah keputusan/aksi yang dilakukan jika kondisi tersebut terpenuhi.

Perlu Anda perhatikan bahwa sebuah "indentasi" diberikan pada kode program setelah *conditional statement* menggunakan kata kunci `if` ditulis. Indentasi diberikan menggunakan "tab" (2 spasi).

```python
# Contoh If (Short Hand)

nilai = 85

if nilai >= 75 : print("Siswa Lulus !")
```

### `if`-`else`

```python
# Contoh If-Else - Contoh 1

nilai = 85

if nilai >= 75 :
  print("Siswa Lulus !")
else :
  print("Siswa Tidak Lulus !")
```

```python
# Contoh If-Else - Contoh 2

nilai = 65

if nilai >= 75 :
  print("Siswa Lulus !")
else :
  print("Siswa Tidak Lulus !")
```

```python
# Contoh If-Else (Short Hand)

nilai = 65

print("Siswa Lulus !") if nilai >= 75 else print("Siswa Tidak Lulus !")
```

### `if`-`elif`-`else`

```python
# Contoh If-Elif-Else - Contoh 1

nilai = 75

nilai_huruf = ''

if nilai >= 85 :
  nilai_huruf = 'A'
elif (nilai >= 75) and (nilai < 85) :
  nilai_huruf = 'B'
elif (nilai >= 50) and (nilai < 75) :
  nilai_huruf = 'C'
else :
  nilai_huruf = 'D'

print('Nilai : ' + str(nilai) + ' (' + nilai_huruf + ')')
```

```python
# Contoh If-Elif-Else - Contoh 2

nilai = 55

nilai_huruf = ''

if nilai >= 85 :
  nilai_huruf = 'A'
elif (nilai >= 75) and (nilai < 85) :
  nilai_huruf = 'B'
elif (nilai >= 50) and (nilai < 75) :
  nilai_huruf = 'C'
else :
  nilai_huruf = 'D'

print('Nilai : ' + str(nilai) + ' (' + nilai_huruf + ')')
```

### Nested `if-else`

```python
# Contoh Nested If-Else - Contoh 1

nilai = 65
jumlah_kehadiran = 15

if nilai >= 75 :
  print("Siswa Lulus !")
else :
  if jumlah_kehadiran >= 10:
    print("Siswa Ikut Ujian Remedy !")
  else :
    print("Siswa Tidak Lulus !")
```

```python
# Contoh Nested If-Else - Contoh 2

nilai = 65
jumlah_kehadiran = 5

if nilai >= 75 :
  print("Siswa Lulus !")
else :
  if jumlah_kehadiran >= 10:
    print("Siswa Ikut Ujian Remedy !")
  else :
    print("Siswa Tidak Lulus !")
```

## *Iterative Statements* (*Loops*)

### `for` *Loop*

Perulangan `for` adalah perulangan yang jumlah perulangannya sudah ditetapkan terlebih dahulu.

```python
# Contoh For Loop - Contoh dengan String

nama = 'Budi'

for huruf in nama :
  print(huruf)
```

```python
# Contoh For Loop - Contoh dengan List

siswa = ['Budi','Ayu','Tono','Putra','Putri']

for nama in siswa :
  print(nama)
```

```python
# Contoh For Loop - Contoh dengan Range - Contoh 1

for index in range(1,11) :
  print(index)
```

```python
# Contoh For Loop - Contoh dengan Range - Contoh 2

for index in range(1,11,2) :
  print(index)
```

### `while` *Loop*

Berbeda dengan perulangan `for`, perulangan `while` adalah perulangan yang jumlah perulangannya tidak ditetapkan terlebih dahulu. Perulangan `while` dijalankan selama suatu kondisi terpenuhi.

```python
# Contoh While Loop - Contoh 1

angka = 1

while angka < 10:
  print(angka)
  angka += 1
```

```python
# Contoh While Loop - Contoh 2 (Contoh Tingkat Lanjut !)

nama = 'Budiman'
huruf_iter = iter(nama)

huruf = next(huruf_iter)
while huruf != 'm' :
  print(huruf)
  huruf = next(huruf_iter)
```

## *Transfer Statements*

### `break`

```python
# Contoh Break

angka = 0

while angka < 5:
    angka += 1
    print(angka)

    if(angka == 3) :
      break

    print('ok')
```

### `continue`

```python
# Contoh Continue

angka = 0

while angka < 5:
    angka += 1
    print(angka)

    if(angka == 3) :
      continue

    print('ok')
```

### `pass`

```python
# Contoh Pass

angka = 0

while angka < 5:
    angka += 1
    print(angka)

    if(angka == 3) :
      pass

    print('ok')
```
